# Polygonal Confinement

### 1) Geometry building

The `ConfiningPolygon` class from the `opensees.section` module makes it convenient to model
Polygonal cross sections that are confined.

```python
import opensees.render.mpl as render
```

```python
from opensees import section
from opensees.section import ConfinedPolygon
```

```python
render.section(ConfinedPolygon(8, 20, s=8));
```

The `s` parameter refines the circular approximation

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2)
render.section(section.ConfiningPolygon(7, 20, 15, s=1), ax=ax[0], show="patch");
render.section(section.ConfiningPolygon(7, 20, 15, s=8), ax=ax[1], show="patch");
```

```python
render.section(section.PolygonRing(10, 20, 18));
```

```python
render.section(section.ConfinedPolygon(8, 20, 20));
```

```python
sect = section.ConfiningPolygon(8, 20, 15, s=8)

ax = render.section(sect, show="patch")

ax.scatter([18, 12],[18, 12])

print((18., 18.) in sect)
print((12, 12) in sect)
```
